## construct a classification model for the iris classification model using logistic regression algorithm 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("C:\\Users\\acer\\Downloads\\iris.csv")
data.head(7)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data["species"].value_counts()   #	sepal_width	petal_length	
```

```python
x = data.drop(columns = ["species"])
y = data["species"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2 , random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression()
```

```python
model = lr.fit(x_train, y_train)
```

```python
model
```

```python
predictions = model.predict(x_test)
predictions.shape
```

```python
result = pd.DataFrame({"actual":y_test, "predicted":predictions})
```

```python
result
```

## model evaluation 

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```
