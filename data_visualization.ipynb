```python
import numpy as np
import matplotlib.pyplot as plt

# Create the face outline
theta = np.linspace(0, 2 * np.pi, 100)
face_x = np.cos(theta)
face_y = np.sin(theta)

# Eyes
left_eye_x = -0.3
left_eye_y = 0.4

right_eye_x = 0.3
right_eye_y = 0.4

# Smile (arc)
smile_theta = np.linspace(0, np.pi, 100)
smile_x = 0.5 * np.cos(smile_theta)
smile_y = -0.5 * np.sin(smile_theta) - 0.2

# Plot
plt.figure(figsize=(6,6))
plt.plot(face_x, face_y, color='gold')  # Face outline
plt.plot(smile_x, smile_y, color='red', linewidth=2)  # Smile

# Eyes
plt.plot(left_eye_x, left_eye_y, 'ko', markersize=10)
plt.plot(right_eye_x, right_eye_y, 'ko', markersize=10)

plt.axis('equal')
plt.axis('off')
plt.title("Smiley Emoji ", fontsize=16)
plt.show()
```
