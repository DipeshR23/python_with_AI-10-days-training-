handling missing values

```python
student = ["muna", np.nan, "ramu" ,np.nan, "kalu"]
age = [21,23,21,30,np.nan]
gender = ["m", "m" ,"m", "f",np.nan]
marks = [23,21,23, 24, 35]
df_std = pd.DataFrame({"student":student, "age":age, "gender":gender, "marks":marks})
df_std
```

```python
df_std.duplicated().sum()
df_std.drop_duplicates(inplace = True)
```

```python
df_std
```

```python
df_std.isnull().sum()    #check null values 
```

techniques to handle missing values 

```python
#remove the row that contains missing vslues
#1 . dropna()
df_std.dropna()
```

###method 2: forward filling and backward filling`

```python
df_std
#fillna()       for fill values
df_std.fillna( method = "ffill")   # for forward fill    
df_std.fillna(method = "bfill")         #for backword fill 
```

```python
df_std.describe()
df_std["gender"].value_counts() #----------- value count 
```

method 3: mean , median and mode imputation

```python
mean_age = df_std["age"].mean()
mean_age
mean_marks = df_std["marks"].mean()
mean_marks
```

```python
gender_mode = df_std["gender"].mode()      #mode return series 
gender_mode = df_std["gender"].mode()[0]     #change series format 
gender_mode  
```

```python
df_filled = df_std.copy()
df_filled["age"] = df_filled["age"].fillna(mean_age)
df_filled
```

```python
df_filled["age"] = df_filled["marks"].fillna(mean_marks)
df_filled
```
